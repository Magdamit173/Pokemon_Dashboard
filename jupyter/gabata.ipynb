```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file
df = pd.read_csv("pokemon.csv")

# Check available columns to make sure "Type 1" exists
print("Columns:", df.columns)

# Count Pokmon by primary type (adjust column name if needed)
type_counts = df["primary_type"].value_counts()


# Create pie chart
plt.figure(figsize=(8, 8))
type_counts.plot.pie(
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,
    wedgeprops={'edgecolor': 'black'}
)
plt.title("Distribution of Pokmon by Type 1")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
